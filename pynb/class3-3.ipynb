```python
# 處理時間:秒數 -> 格林威治時間
from datetime import datetime
str(datetime.utcfromtimestamp(1140825600))
```

```python
# 轉換流程->遞給整行->每一個走過流程
def timetrans(s):
    return str(datetime.utcfromtimestamp(s))
# apply: apply(流程名字) 對每一個執行一次流程
df["film_date"] = df["film_date"].apply(timetrans)
df
```

```python
# 篩選操作: 原理 df[[跟你的列數一樣多的布林]]
df.head(3)[[True,False,True]]
```

```python
#篩選條件
def fflow(n):
    if n > 3000 and n < 5000:
        return True
    else:
        return False
f = df["comments"].apply(fflow)
df[f]
```
